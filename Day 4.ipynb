## Descriptive Statistic (Summarizing)

Decriptive statistics are numbers that are used to describe or summarizes data, and typically is the first kind of analysis performed on dataset.

Descriptive statistics are broken down into measures of **central tendency** and **measures of variability, or spread**. 

## Central Tendency

Central tendency is a measures of central location of the distribution.
The most common measures of central tendency are **the mean**, **the median** and **the mode**.

**The Mean** is the **average**, The sum of a set of data divided by the number of data.

**The Median** is The **middle value** when data is arranged in numerical order.

**The Mode** is the value that **appeared the most**

## So which measure of central tendency to use?

**The Mean**, Use the mean to describe the middle of a set of data that **does not have an outlier**.

**The Median**, Use the median to describe the middle of a set of data that **does have an outlier**.

**The Mode**, Use the mode when the data is **non-numeric** or when asked to choose the **most popular item**.

```r
getmode = function(v) {
   uniqv = unique(v)
   uniqv[which.max(tabulate(match(v, uniqv)))]
}
```

```r
 N = 10000
 x = round(rnorm(N,mean=10, sd=2))
 hist(x, 
 xlim=c(min(x),max(x)), probability=T, nclass=max(x)-min(x)+1, 
   col='white', xlab=' ', ylab=' ', axes=F,
   main='Normal Distribution')
#lines(density(x,bw=1), col='', lwd=2)
abline(v=mean(x), col = "blue", lwd = 2)
abline(v=median(x), col = "red", lwd = 2)
abline(v=getmode(x), col = "green", lwd = 2)

legend("topright", c("Mean", "Median","Mode"), col=c("blue", "red","Green"), lwd=2)
```

```r
 N = 10000
 x = rnbinom(N, 3, .2)
 hist(x, 
 xlim=c(min(x),max(x)), probability=T, nclass=max(x)-min(x)+1, 
   col='white', xlab=' ', ylab=' ', axes=F,
   main='Positive Skewed')
#lines(density(x,bw=1), col='', lwd=2)
abline(v=mean(x), col = "blue", lwd = 2)
abline(v=median(x), col = "red", lwd = 2)
abline(v=getmode(x), col = "green", lwd = 2)

legend("topright", c("Mean", "Median","Mode"), col=c("blue", "red","Green"), lwd=2)
```

## Measure of Variability

Measure of Variability is a measure of the **spread of a data set**, Measures of **variability** includes the **standard deviation or variance, the minimum and maximum variables, and skewness**.

**Standard Deviation** is a measure of how **spread out** numbers are

**Variance** is the average of the squared differences from the mean

```r
x = rnorm(10, mean = 5,sd = 1)
x
mean_x = mean(x)
mean_x

#difference from mean
diff = (x-mean_x)^2
diff
variance = sum(diff/(length(diff)-1))
variance
std = sqrt(variance)
std

#result using sd()
sd(x)
```

# {base} Graphics System
- Common main plotting functions: **hist()**, **barplot()**, **boxplot()**, **plot()**
- Annotation: **points()**, **abline()**, **text()**
- Parameters:
    - pch: the **plotting symbol** (default is open circle)
    - lty: the **line type** (default is solid line)
    - lwd: the **line width** 
    - col: the plotting **color**, specified as a number, string, or hex code
    - xlab: character string for the x-axis label
    - ylab: character string for the y-axis
- Global Parameters: **par()**
    - las: the orientation of the axis labels on the plot
    - bg: the background color
    - mar: the margin size
    - oma: the outer margin size (default is 0 for all sides)
    - mfrow: number of plots per row, column (plots are filled row-wise)
    - mfcol: number of plots per row, column (plots are filled column-wise)

```r
par(mar = rep(2,4),mfrow=c(3,1))
df = data.frame(x = c(3,1,5), y = c(2,4,6), label = c('a','b','c'))
p = ggplot(df, aes(x,y,label = label)) + xlab(NULL) + ylab(NULL) 
p + geom_point() + ggtitle('geom_point') 
p + geom_bar(stat = 'identity') +ggtitle("geom_bar(stat=\"identity\")")
p + geom_line() + ggtitle("geom_line")
p + geom_area() + ggtitle("geom_area")
p + geom_path() + ggtitle("geom_path")
p + geom_text() + ggtitle("geom_text")
p + geom_tile() + ggtitle("geom_tile")
p + geom_polygon() + ggtitle("geom_polygon")
```

## Position Adjustment
- **dodge** = Adjust position by dodging overlaps to the side
- **fill** = Stack overlapping objects and standardise have equal height
- **identity** = Dont adjust position
- **jitter** = Jitter points to avoid overplotting
- **stack** = Stack overlapping object on top of one another

## Line Charts
